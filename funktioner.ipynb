# Funktioner

* inparameter
* anropa funktion
* return sats
* lokala variabler

```python
# definerat en funktion som returnerar medelvärdet av två tal
def medelvärde(tal1,tal2):
    medel=(tal1+tal2)/2 # medel är en lokal avriabel - syns endast i scopet av funktionen
    return medel

vikt_medel=medelvärde(60,62)

print(f"Min medelvikt är {vikt_medel} kg")
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return x**2

# anropa funktionen med parametern 5
print(f(5))

x=np.linspace(-10,10)
y=f(x)
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("F(x)=x^2")
plt.grid()
plt.show()
```
